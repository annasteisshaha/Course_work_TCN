```python
from transformers import pipeline
import pandas as pd
import numpy as np
import torch as th
```

```python
pipe = pipeline("feature-extraction", model="DeepPavlov/rubert-base-cased")
```

```python
data = pd.read_excel('/content/Оквэды с инфо.xlsx', index_col=0)
data.head()
```

```python
embeddings_full = pipe(data['info'].values.tolist())
```

```python
lst = []

for i in embeddings_full:
  lst.append(np.array(i[0]).mean(axis=0).tolist())
```

```python
embeddings_okved = np.array(lst)
np.save('Embeddings_okved.npy', embeddings_okved)
```
